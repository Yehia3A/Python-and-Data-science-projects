```python
from ultralytics import YOLO

# Load the model configuration
model = YOLO('ultralytics/cfg/models/12/yolo12.yaml')

results = model.train(
    data='/kaggle/input/cardetection/car/data.yaml',
    epochs=10,  # Increase for better convergence
    batch=16,
    imgsz=640,
    scale=0.5,
    mosaic=1.0,
    flipud=0.5,  # Flipping
    hsv_h=0.015,  # Hue shift (brightness)
    hsv_s=0.7,    # Saturation shift
    hsv_v=0.4,    # Value shift
    copy_paste=0.1,
    device=0,
    name="cardetection",
    cls=0.5  # Class-weighted loss for imbalance
)
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
import os

# Load the trained YOLOv12 model
model = YOLO('runs/detect/cardetection/weights/best.pt')  # Path to your trained model

# Print model classes to verify
print("Model classes:", model.names)

# Video path
video_path = "/kaggle/input/cardetection/video.mp4"

# Open the video
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get video properties
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(5))

# Define output paths
raw_output_path = "/kaggle/working/raw_detected_video.mp4"
filtered_output_path = "/kaggle/working/filtered_detected_video.mp4"
log_file = "/kaggle/working/detection_log.txt"
sample_frame_path = "/kaggle/working/sample_detected_frame.jpg"

# Initialize video writers
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
raw_out = cv2.VideoWriter(raw_output_path, fourcc, fps, (frame_width, frame_height))
filtered_out = cv2.VideoWriter(filtered_output_path, fourcc, fps, (frame_width, frame_height))

# Initialize log and metrics
log = []
false_positives_raw = 0
false_positives_filtered = 0
sample_frame_saved = False

# Process video frames
frame_count = 0
# Process video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame_count += 1
    results = model(frame, conf=0.25)  # Standard confidence threshold
    raw_frame = frame.copy()
    for r in results:
        boxes = r.boxes.xyxy.cpu().numpy()
        scores = r.boxes.conf.cpu().numpy()
        classes = r.boxes.cls.cpu().numpy()
        names = r.names
        for box, score, cls in zip(boxes, scores, classes):
            x1, y1, x2, y2 = map(int, box)
            label = names[int(cls)]
            cv2.rectangle(raw_frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
            cv2.putText(raw_frame, f"{label}: {score:.2f}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
    raw_out.write(raw_frame)
    if not sample_frame_saved and len(boxes) > 0:
        cv2.imwrite(sample_frame_path, raw_frame)
        sample_frame_saved = True
# Release resources
cap.release()
raw_out.release()
filtered_out.release()

# Save log
with open(log_file, 'w') as f:
    f.write("\n".join(log))

# Print metrics
false_positive_reduction = false_positives_raw - false_positives_filtered
print(f"Raw detection video saved to {raw_output_path}")
print(f"Filtered detection video saved to {filtered_output_path}")
print(f"Log saved to {log_file}")
print(f"Metrics: False positives reduced from {false_positives_raw} to {false_positives_filtered} "
      f"({false_positive_reduction} removed)")
```

```python
# Bonus Task: Track and Ignore Adversarial Pedestrians Wearing Stop Signs
from deep_sort_realtime.deepsort_tracker import DeepSort
import numpy as np
import cv2
from ultralytics import YOLO

# Load the trained model
model = YOLO('runs/detect/cardetection/weights/best.pt')
```

```python

# Initialize DeepSORT tracker
tracker = DeepSort(max_age=30, n_init=3, nms_max_overlap=1.0, max_cosine_distance=0.2)

# Paths for input and output videos
video_path = "/kaggle/input/videom3/videom3.mp4"
raw_output_path = "/kaggle/working/raw_detected_video.mp4"
filtered_output_path = "/kaggle/working/filtered_detected_video.mp4"

# Open video
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get video properties
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(5))

# Initialize video writers
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
raw_out = cv2.VideoWriter(raw_output_path, fourcc, fps, (frame_width, frame_height))
filtered_out = cv2.VideoWriter(filtered_output_path, fourcc, fps, (frame_width, frame_height))

# Log for removed detections
removed_detections_log = []

# Process video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run object detection
    results = model(frame)
    detections = []
    for r in results:
        boxes = r.boxes.xyxy.cpu().numpy()
        scores = r.boxes.conf.cpu().numpy()
        classes = r.boxes.cls.cpu().numpy()
        for box, score, cls in zip(boxes, scores, classes):
            x1, y1, x2, y2 = box
            label = r.names[int(cls)]
            if label in ['person', 'stop sign']:
                detections.append([x1, y1, x2, y2, score, label])

    # Convert detections to DeepSORT format
    dets = []
    labels = []
    for det in detections:
        x1, y1, x2, y2, score, label = det
        dets.append([x1, y1, x2 - x1, y2 - y1, score])
        labels.append(label)
    dets = np.array(dets)

    # Update tracker with detections
    tracks = tracker.update_tracks(dets, frame=frame)

    # Process tracks and filter stop signs
    raw_annotated_frame = frame.copy()
    filtered_annotated_frame = frame.copy()
    person_boxes = []
    stop_sign_boxes = []

    for track in tracks:
        if not track.is_confirmed():
            continue
        x1, y1, x2, y2 = track.to_tlbr()
        label = labels[track.detection_index]
        if label == 'person':
            person_boxes.append([x1, y1, x2, y2])
            cv2.rectangle(raw_annotated_frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(raw_annotated_frame, 'person', (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        elif label == 'stop sign':
            stop_sign_boxes.append([x1, y1, x2, y2])
            cv2.rectangle(raw_annotated_frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 0, 255), 2)
            cv2.putText(raw_annotated_frame, 'stop sign', (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

    # Filter stop signs inside person boxes using IoU
    def compute_iou(boxA, boxB):
        xA = max(boxA[0], boxB[0])
        yA = max(boxA[1], boxB[1])
        xB = min(boxA[2], boxB[2])
        yB = min(boxA[3], boxB[3])
        interArea = max(0, xB - xA) * max(0, yB - yA)
        boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
        boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
        iou = interArea / float(boxAArea + boxBArea - interArea)
        return iou

    for stop_box in stop_sign_boxes:
        removed = False
        for person_box in person_boxes:
            iou = compute_iou(stop_box, person_box)
            if iou > 0.5:
                removed_detections_log.append(f"Frame {cap.get(1)}: Stop sign at {stop_box} removed due to overlap with person at {person_box}, IoU={iou:.2f}")
                removed = True
                break
        if not removed:
            cv2.rectangle(filtered_annotated_frame, (int(stop_box[0]), int(stop_box[1])), (int(stop_box[2]), int(stop_box[3])), (0, 0, 255), 2)
            cv2.putText(filtered_annotated_frame, 'stop sign', (int(stop_box[0]), int(stop_box[1]) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

    # Draw person boxes on filtered frame
    for person_box in person_boxes:
        cv2.rectangle(filtered_annotated_frame, (int(person_box[0]), int(person_box[1])), (int(person_box[2]), int(person_box[3])), (0, 255, 0), 2)
        cv2.putText(filtered_annotated_frame, 'person', (int(person_box[0]), int(person_box[1]) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Write frames to output videos
    raw_out.write(raw_annotated_frame)
    filtered_out.write(filtered_annotated_frame)

# Release resources
cap.release()
raw_out.release()
filtered_out.release()
```

```python


# Save and display log
with open('/kaggle/working/removed_detections_log.txt', 'w') as f:
    f.write('\n'.join(removed_detections_log))
print("Removed detections log saved to /kaggle/working/removed_detections_log.txt")
```

```python

# Calculate false positive reduction
initial_stop_signs = sum(1 for det in detections if det[5] == 'stop sign')
filtered_stop_signs = len(stop_sign_boxes) - len(removed_detections_log)
false_positive_reduction = ((initial_stop_signs - filtered_stop_signs) / initial_stop_signs * 100) if initial_stop_signs > 0 else 0
print(f"Initial stop signs detected: {initial_stop_signs}")
print(f"Filtered stop signs: {filtered_stop_signs}")
print(f"False positive reduction: {false_positive_reduction:.2f}%")

```
